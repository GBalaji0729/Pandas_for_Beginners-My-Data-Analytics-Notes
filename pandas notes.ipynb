# 3. Merge, join, and concatenate.

- how=’inner‘:will return the common elements in the two data frames.
- how=’outer‘:will merge two data frames without duplicates.
- how=’left’: will merge the entire left data frame with the same data in the right data frame without duplicates.
- how=’right’:will merge the entire right data with the same data in the left data frame with out duplicates.

## 🔹 a. `merge()`

- Works like **SQL joins** (INNER, LEFT, RIGHT, OUTER).
- You specify the **common column(s)** (`on`, `left_on`, `right_on`).
- Used when you want to combine **rows based on keys**.

## 🔹 b. `join()`

- Similar to `merge`, but usually works on **index** instead of columns.
- By default, it joins on the **row index**.
- Convenient when you already have **matching indexes**

## 🔹 c. `concat()`

- Used to **stack DataFrames** either **vertically (rows)** or **horizontally (columns)**.
- Doesn’t look for matching keys unless you tell it (`axis`, `ignore_index`).
- Think of it like **appending or combining datasets**.

# 4. Series
A Series is a one-dimensional labeled array in pandas that can hold data of any type (integers, strings, floats, objects, etc.).

- Each element in a Series has a label (index) and a value.
- It is similar to a single column of a DataFrame or a list with labels

differences between Series and DataFrame:


| Feature   | Series (1D)               | DataFrame (2D)                 |
| --------- | ------------------------- | ------------------------------ |

# 6. Combining and concating Series 


**1. What does "combine" mean?**

In pandas, combining two Series means taking them element by element and deciding, with a function, how to merge the two values.
It’s not stacking like concat — instead, it’s pairwise merging.

**working of lambda x, y: max (x ,y) :**

series1 = [10, 25, 5, 30, 15],
series2 = [5, 30, 10, 15, 40]

- Pair 1: x=10, y=5 → max(10,5)=10

- Pair 2: x=25, y=30 → max(25,30)=30

- Pair 3: x=5, y=10 → max(5,10)=10

- Pair 4: x=30, y=15 → max(30,15)=30

- Pair 5: x=15, y=40 → max(15,40)=40

**2. What is concat?**

pd.concat() is used to combine/stack multiple pandas objects (Series or DataFrames) along a particular axis (rows or columns)

**Key things to notice:**

- The values of s1 come first, then the values of s2.

- The indexes are preserved, so you see two 0s, two 1s, two 2s.
(Indexes are not reset automatically)

```python
import pandas as pd

data1 = [10, 25, 5, 30, 15]
data2 = [5, 30, 10, 15, 40]

series1 = pd.Series(data1)
series2 = pd.Series(data2)

db=series1.combine(series2,lambda x, y: max (x ,y))
print(db)

# lambda as a short cut to write function
# eg: def demo(x, y):
#          return max(x,y)
```

```python
data1 = [10, 25, 5, 30, 15]
data2 = [5, 30, 10, 15, 40]

series1 = pd.Series(data1)
series2 = pd.Series(data2)

result=pd.concat([series1, series2], ignore_index=True)
print(result)
# ignore_index=True will reset the index of the resulting Series after concatenation.
```

# 8. Cut and Qcut in categorical data.

**pd.cut:**   cut() divides continuous numeric data into equal-width bins (intervals) that you define.

**pd.qcut:**  qcut() divides data into equal-sized groups (quantiles) based on the distribution of the data — not fixed values.

```python
data = pd.DataFrame({"Age": [5, 12, 17, 24, 35, 46, 59, 70, 85]})

# Create bins
bins = [0, 12, 19, 35, 60, 100]
labels = ["Child", "Teen", "Young Adult", "Adult", "Senior"]

# Apply the bins to the data
data["Agegroup"]=pd.cut(data["Age"] ,bins=bins , labels=labels)
print(data)
```

```python
# append new category

s=pd.Series(
    data=["a", "b", "c", "d", "e"],dtype="category"
    
)

print(s)
s=s.cat.add_categories("f")

print(s)
```

### filter() is used to select specific rows or columns based on:

- **items (Exact Names):**

  Definition: Selects rows or columns that exactly match the given label names.

  Use case: When you know the exact column or index names you want to keep 
   


- **like (Partial Matches):**

  Definition: Selects rows or columns whose names contain a given substring.

  Use case: When you want to extract multiple columns with similar names



- **regex (Regular Expressions):**

  Definition: Uses a regular expression pattern to match rows or columns.

  Use case: When you need advanced or pattern-based filtering (e.g., all columns starting or ending with certain words or numbers).

  
  

### The axis parameter :

The axis parameter tells pandas whether you’re filtering rows or columns.

```python
pf=df2.set_index("Country")
pf
```

```python
pf.filter(items=["Continent","CCA3"],axis=1) #axis=1 acts on columns || axis=0 acts on Rows in this case when no rows named with "Continent"and "CCA3" are found hence it returns empty dataframe 
                                            
```

```python
pf.filter(like="United",axis=0)
```

```python
# this will display the columns headders starts with 202

pf.filter(regex="^202",axis=1)
```

# 12. Groupby and Aggregating in Pandas

```python

data = {
    'Department': ['Sales', 'Sales', 'HR', 'HR', 'IT', 'IT'],
    'Employee': ['A', 'B', 'C', 'D', 'E', 'F'],
    'Salary': [50000, 55000, 60000, 58000, 65000, 62000]
}

df = pd.DataFrame(data)
print(df)
```

```python
grouped=df.groupby("Department")
grouped
```

```python
grouped["Employee"].count()
```

```python
grouped["Salary"].sum()
```

```python
grouped['Salary'].agg(['sum', 'mean', 'max', 'min'])
```

```python
#it gives min and max salary department wise 

df.groupby('Department')['Salary'].agg(['min', 'max'])
```
